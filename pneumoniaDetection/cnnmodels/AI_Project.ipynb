```python
# let's build the CNN model

cnn = Sequential()

#Convolution layer 1
cnn.add(Conv2D(32, (3, 3), activation="relu", input_shape=(64, 64, 3)))

#Pooling
cnn.add(MaxPooling2D(pool_size = (2, 2)))

# 2nd Convolution
cnn.add(Conv2D(32, (3, 3), activation="relu"))

# 2nd Pooling layer
cnn.add(MaxPooling2D(pool_size = (2, 2)))

# Flatten the layer
cnn.add(Flatten())

# Connected Layers
cnn.add(Dense(activation = 'relu', units = 128))
cnn.add(Dense(activation = 'sigmoid', units = 1))

# Compiling the Neural network
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
num_of_test_samples = 600
batch_size = 25
```

```python
# Fitting the CNN to the images
# The function ImageDataGenerator augments your image by iterating through image as your CNN is getting ready to process the image

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)  #Image normalization.

training_set = train_datagen.flow_from_directory(r'C:\Users\Bhuvan Suresh\Desktop\AI Project\pneumoniaDetection\cnnmodels\dataset\chest_xray\train',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')

validation_generator = test_datagen.flow_from_directory(r'C:\Users\Bhuvan Suresh\Desktop\AI Project\pneumoniaDetection\cnnmodels\dataset\chest_xray\val',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary')

test_set = test_datagen.flow_from_directory(r'C:\Users\Bhuvan Suresh\Desktop\AI Project\pneumoniaDetection\cnnmodels\dataset\chest_xray\test',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

```python
#the contents of our model
cnn.summary()
```

```python
#fitting the data into training and validation data set
cnn_model = cnn.fit(training_set,
                         steps_per_epoch = 16,
                         epochs = 160,
                         validation_data = validation_generator,
                         validation_steps = 16)
```

```python
# how good is our model with test data?
test_accu = cnn.evaluate(test_set,steps= 39)
```

```python
print('The testing accuracy is :',test_accu[1]*100, '%')
```

```python
#efficient way 1
#to predict new images 
import keras.utils as image
def predict_image(imagepath, cnn):
    predict = image.load_img(imagepath, target_size = (64, 64))   
    predict_modified = image.img_to_array(predict)
    predict_modified = predict_modified / 255
    predict_modified = np.expand_dims(predict_modified, axis = 0)
    result = cnn.predict(predict_modified)
    if result[0][0] >= 0.5:
        prediction = 'pneumoic'
        probability = result[0][0]
        print ("probability = " + str(probability))
    else:
        prediction = 'normal'
        probability = 1 - result[0][0]
        print ("probability = " + str(probability))
    print("Prediction = " + prediction)
```
